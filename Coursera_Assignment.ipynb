```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load Boston dataset from a CSV file
url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
data = pd.read_csv(url)

# Create the boxplot for the median value of owner-occupied homes (MEDV)
def boxplot_medv():
    plt.figure(figsize=(8, 6))
    sns.boxplot(y=data['medv'])
    plt.title("Boxplot for Median Value of Owner-Occupied Homes (MEDV)")
    plt.ylabel("Median Value (in $1000s)")
    plt.show()

# Create the barplot for the Charles River variable (chas)
def barplot_chas():
    plt.figure(figsize=(8, 6))
    sns.countplot(x=data['chas'])
    plt.title("Barplot for Charles River Variable (chas)")
    plt.xlabel("CHAS (0 = No, 1 = Yes)")
    plt.ylabel("Count")
    plt.show()

# Create the boxplot for MEDV vs AGE
def boxplot_medv_age():
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=pd.cut(data['age'], bins=5), y=data['medv'])
    plt.title("Boxplot for MEDV vs AGE")
    plt.xlabel("Age Groups")
    plt.ylabel("Median Value (in $1000s)")
    plt.xticks(rotation=45)
    plt.show()

# Create the scatter plot for NOX vs INDUS
def scatter_nox_indus():
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=data['nox'], y=data['indus'])
    plt.title("Scatter Plot of Nitric Oxide Concentrations (NOX) vs Proportion of Non-Retail Business Acres (INDUS)")
    plt.xlabel("Nitric Oxide Concentration (NOX)")
    plt.ylabel("Proportion of Non-Retail Business Acres (INDUS)")
    plt.show()

# Create the histogram for the pupil-teacher ratio (ptratio)
def histogram_ptratio():
    plt.figure(figsize=(8, 6))
    sns.histplot(data['ptratio'], kde=True)
    plt.title("Histogram for Pupil-Teacher Ratio (PTRATIO)")
    plt.xlabel("Pupil-Teacher Ratio")
    plt.ylabel("Frequency")
    plt.show()

# Call the functions to display the plots
if __name__ == "__main__":
    print("Generating plots...\n")
    
    # Generate each plot
    boxplot_medv()
    barplot_chas()
    boxplot_medv_age()
    scatter_nox_indus()
    histogram_ptratio()
```
